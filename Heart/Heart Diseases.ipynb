```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Heart.csv')
```

```python
df.head()
```

```python
df.shape
```

### Data Cleaning


```python
df.drop(columns = {'Unnamed: 0'}, inplace = True)
```

### Convert all not numeric columns to numeric

```python
df['ChestPain'] = df['ChestPain'].astype('category') 
df['ChestPain']  = df['ChestPain'].cat.codes
```

```python
df['Thal'] = df['Thal'].astype('category') 
df['Thal']  = df['Thal'].cat.codes
```

```python
df['AHD'] = df['AHD'].astype('category') 
df['AHD']  = df['AHD'].cat.codes
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=  True)
```

```python
df.duplicated().sum()
```

### Independent variable X 


```python
X = df.drop(columns = {'AHD'}, axis = 1)
X.head()
```

### Dependent variable Y 


```python
Y = df.AHD
Y.head()
```

### Split data into train and test set

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = .3, random_state = 21)
```

```python
x_train
```

```python
x_test
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_train_scaled
```

```python
x_test_scaled = scaler.transform(x_test)
x_test_scaled
```

```python
model = linear_model.LogisticRegression(random_state= 0 ).fit(x_train_scaled, y_train)
```

```python
model.predict(x_train_scaled)
```

```python
model.score(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled, y_test)
```
